```python
import pandas as pd

def trim_csv(file_path, num_rows, output_file_path):
    """
    Fungsi ini akan memangkas jumlah record pada sebuah file CSV.

    Parameters:
    file_path (str): Lokasi dari file CSV.
    num_rows (int): Jumlah baris yang ingin dipertahankan.
    output_file_path (str): Lokasi output file CSV yang sudah dipangkas.

    Returns:
    None
    """
    # Membaca file CSV
    df = pd.read_csv(file_path)

    # Memangkas DataFrame
    df_trimmed = df.head(num_rows)

    # Menyimpan DataFrame yang sudah dipangkas ke dalam file CSV baru
    df_trimmed.to_csv(output_file_path, index=False)

# Contoh penggunaan fungsi
trim_csv('/content/drive/MyDrive/nlp-project/translated_file.csv', 5000, '/content/drive/MyDrive/nlp-project/5000-Data.csv')
```
