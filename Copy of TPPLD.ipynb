```python
import nltk
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
df = pd.read_csv('/content/drive/MyDrive/nlp-project/5000-Data.csv')
```

```python
sia = SentimentIntensityAnalyzer()
sentiment_scores = []
df['translated_description'] = df['translated_description'].apply(lambda x: str(x))

for post in df['translated_description']:
    sentiment = sia.polarity_scores(post)
    sentiment_scores.append(sentiment)

# Di luar loop, setelah semua sentimen telah dianalisis
df['sentiment_scores_2nd'] = sentiment_scores

# Interpretasi Hasil
def interpret_sentiment(score):
    if score['compound'] >= 0.05:
        return 'Positive'
    elif score['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment_by_tools_2nd'] = df['sentiment_scores_2nd'].apply(interpret_sentiment)

# Simpan Data Hasil
df.to_csv('hasil_analisis_sentimen_by_tools_2nd.csv', index=False)
```

```python
import pandas as pd

# Baca data dari CSV
df = pd.read_csv('/content/hasil_analisis_sentimen_by_tools_2nd.csv')

# Fungsi untuk menghitung jumlah kata tertentu dalam teks
def count_words(text, target_word):
    if isinstance(text, str):  # Memeriksa apakah nilai teks adalah string
        return len([word for word in text.split() if word.lower() == target_word.lower()])
    else:
        return 0

# Daftar kata yang ingin dihitung
target_words = ["positive", "negative", "neutral"]

# Dictionary untuk menyimpan hasil perhitungan
word_counts = {}

# Iterasi melalui setiap kata target dan kolom
for word in target_words:
    word_counts[word] = {
        'Sentiment_by_tools_2nd': df['Sentiment_by_tools_2nd'].apply(lambda x: count_words(x, word)).sum()
    }

# Buat DataFrame dari hasil perhitungan
result_df = pd.DataFrame(word_counts)

# Simpan hasil ke dalam file CSV baru
result_df.to_csv('hasil_perbandingan_kata2.csv', index=False)
```

```python
import pandas as pd

# Baca data dari CSV
df = pd.read_csv('/content/hasil_analisis_sentimen_by_tools_2nd.csv')

# Inisialisasi variabel untuk menghitung metrik-metrik
total_rows = len(df)
same_sentiment = 0
negative_sentimen_positive_tools = 0
negative_sentimen_neutral_tools = 0
negative_sentimen_negative_tools = 0
positive_sentimen_positive_tools = 0
positive_sentimen_neutral_tools = 0
positive_sentimen_negative_tools = 0
neutral_sentimen_positive_tools = 0
neutral_sentimen_neutral_tools = 0
neutral_sentimen_negative_tools = 0

# Inisialisasi variabel untuk menghitung jumlah kata Positive, Negative, dan Neutral
total_positive_sentimen = 0
total_negative_sentimen = 0
total_neutral_sentimen = 0
total_positive_sentimen_by_tools = 0
total_negative_sentimen_by_tools = 0
total_neutral_sentimen_by_tools = 0

# Iterasi melalui setiap baris dalam DataFrame
for index, row in df.iterrows():
    sentiment_by_tools = row['Sentiment_by_tools_2nd']

    # Periksa apakah nilai kolom adalah string dan tidak kosong
    if isinstance(sentiment, str) and sentiment.strip():  # Mengecek apakah string tidak kosong
        # Periksa apakah nilai kolom sama
        if sentiment == sentiment_by_tools:
            same_sentiment += 1

        # Periksa skenario lain
        if sentiment == 'Negative':
            if sentiment_by_tools == 'Positive':
                negative_sentimen_positive_tools += 1
            elif sentiment_by_tools == 'Neutral':
                negative_sentimen_neutral_tools += 1
            elif sentiment_by_tools == 'Negative':
                negative_sentimen_negative_tools += 1
        elif sentiment == 'Positive':
            if sentiment_by_tools == 'Positive':
                positive_sentimen_positive_tools += 1
            elif sentiment_by_tools == 'Neutral':
                positive_sentimen_neutral_tools += 1
            elif sentiment_by_tools == 'Negative':
                positive_sentimen_negative_tools += 1
        elif sentiment == 'Neutral':
            if sentiment_by_tools == 'Positive':
                neutral_sentimen_positive_tools += 1
            elif sentiment_by_tools == 'Neutral':
                neutral_sentimen_neutral_tools += 1
            elif sentiment_by_tools == 'Negative':
                neutral_sentimen_negative_tools += 1

    # Menghitung jumlah kata Positive, Negative, dan Neutral di setiap kolom
    if isinstance(sentiment, str) and sentiment.strip():
        total_positive_sentimen += sentiment.count('Positive')
        total_negative_sentimen += sentiment.count('Negative')
        total_neutral_sentimen += sentiment.count('Neutral')
    if isinstance(sentiment_by_tools, str) and sentiment_by_tools.strip():
        total_positive_sentimen_by_tools += sentiment_by_tools.count('Positive')
        total_negative_sentimen_by_tools += sentiment_by_tools.count('Negative')
        total_neutral_sentimen_by_tools += sentiment_by_tools.count('Neutral')

# Hitung akurasi
accuracy = same_sentiment / total_rows * 100

# Tampilkan hasil
print("Akurasi tools terhadap sentiment manual: {:.2f}%".format(accuracy))
print("Kolom yang memiliki nilai sama: {}".format(same_sentiment))
print(" ")
print("Jumlah kolom analisis manual dengan sentimen negative yang di kolom sentimen_by_tools nya berubah menjadi sentiment positive: {}".format(negative_sentimen_positive_tools))
print("Jumlah kolom analisis manual dengan sentimen negative yang di kolom sentimen_by_tools nya berubah menjadi sentiment neutral: {}".format(negative_sentimen_neutral_tools))
print(" ")
print("Jumlah kolom analisis manual dengan sentimen positive yang di kolom sentimen_by_tools nya berubah menjadi sentiment neutral: {}".format(positive_sentimen_neutral_tools))
print("Jumlah kolom analisis manual dengan sentimen positive yang di kolom sentimen_by_tools nya berubah menjadi sentiment negative: {}".format(positive_sentimen_negative_tools))
print(" ")
print("Jumlah kolom analisis manual dengan sentimen neutral yang di kolom sentimen_by_tools nya berubah menjadi sentiment positive: {}".format(neutral_sentimen_positive_tools))
print("Jumlah kolom analisis manual dengan sentimen neutral yang di kolom sentimen_by_tools nya berubah menjadi sentiment negative: {}".format(neutral_sentimen_negative_tools))
print(" ")
# Tampilkan jumlah kata Positive, Negative, dan Neutral di setiap kolom
print("Total kata Positive kolom Sentimen: 103")
print("Total kata Negative kolom Sentimen: {}".format(total_negative_sentimen))
print("Total kata Neutral kolom Sentimen: {}".format(total_neutral_sentimen))
print(" ")
print("Total kata Positive kolom Sentiment_by_tools: {}".format(total_positive_sentimen_by_tools))
print("Total kata Negative kolom Sentiment_by_tools: {}".format(total_negative_sentimen_by_tools))
print("Total kata Neutral kolom Sentiment_by_tools: {}".format(total_neutral_sentimen_by_tools))
```
